```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import textwrap
from sklearn.preprocessing import LabelEncoder

# ===============================
# Assume 'df' is your cleaned dataset
# ===============================

# -------------------------------
# 1️⃣ Keep numeric copies for ML
# -------------------------------
# Encode categorical columns for ML
cat_cols = ['Brand', 'Processor_Brand', 'GPU_Brand', 'Display_type']
le_dict = {}

for col in cat_cols:
    le = LabelEncoder()
    df[col + '_encoded'] = le.fit_transform(df[col])
    le_dict[col] = le  # store encoder for reference

# -------------------------------
# 2️⃣ Convert numeric strings to int
# -------------------------------
numeric_cols = ['RAM', 'SSD']
for col in numeric_cols:
    df[col] = df[col].astype(str).str.extract(r'(\d+)')[0].fillna(0).astype(int)

# If RAM_Expandable exists
if 'RAM_Expandable' in df.columns:
    df['RAM_Expandable'] = df['RAM_Expandable'].astype(str).str.extract(r'(\d+)')[0].fillna(0).astype(int)

# -------------------------------
# 3️⃣ Helper function to wrap long labels
# -------------------------------
def wrap_labels(col, width=12):
    return col.apply(lambda x: '\n'.join(textwrap.wrap(str(x), width)))

# -------------------------------
# 4️⃣ Polished Plots
# -------------------------------
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12,6)

# 1️⃣ Price Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Price'], kde=True, bins=30, color='skyblue')
plt.title('Laptop Price Distribution', fontsize=16)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(y=df['Price'], color='lightgreen')
plt.title('Price Boxplot (Outliers)', fontsize=16)
plt.ylabel('Price', fontsize=12)
plt.tight_layout()
plt.show()

# 2️⃣ Average Price by Brand
plt.figure(figsize=(12,6))
sns.barplot(x='Brand', y='Price', data=df, palette='Set2')
plt.xticks(rotation=45, ha='right')
plt.title('Average Price by Brand', fontsize=16)
plt.xlabel('Brand', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.tight_layout()
plt.show()

# 3️⃣ RAM vs Price
plt.figure(figsize=(10,6))
sns.scatterplot(x='RAM', y='Price', hue='Brand', data=df, s=100, palette='Set1')
plt.title('RAM vs Price by Brand', fontsize=16)
plt.xlabel('RAM (GB)', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 4️⃣ SSD vs Price
plt.figure(figsize=(10,6))
sns.scatterplot(x='SSD', y='Price', hue='Brand', data=df, s=100, palette='Set1')
plt.title('SSD vs Price by Brand', fontsize=16)
plt.xlabel('SSD (GB)', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 5️⃣ Processor Brand vs Price
df['Processor_Brand_wrapped'] = wrap_labels(df['Processor_Brand'], width=12)
plt.figure(figsize=(14,6))
sns.boxplot(x='Processor_Brand_wrapped', y='Price', data=df, palette='pastel')
plt.xticks(rotation=60, ha='right')
plt.title('Processor Brand vs Price', fontsize=16)
plt.xlabel('Processor Brand', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.tight_layout()
plt.show()

# 6️⃣ GPU Brand vs Price
df['GPU_Brand_wrapped'] = wrap_labels(df['GPU_Brand'], width=12)
plt.figure(figsize=(14,6))
sns.boxplot(x='GPU_Brand_wrapped', y='Price', data=df, palette='coolwarm')
plt.xticks(rotation=60, ha='right')
plt.title('GPU Brand vs Price', fontsize=16)
plt.xlabel('GPU Brand', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.tight_layout()
plt.show()

# 7️⃣ Display Type vs Price
df['Display_type_wrapped'] = wrap_labels(df['Display_type'], width=12)
plt.figure(figsize=(12,6))
sns.boxplot(x='Display_type_wrapped', y='Price', data=df, palette='Set3')
plt.xticks(rotation=45, ha='right')
plt.title('Display Type vs Price', fontsize=16)
plt.xlabel('Display Type', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.tight_layout()
plt.show()

# 8️⃣ Correlation Heatmap (numeric columns only)
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='vlag', fmt=".2f", linewidths=0.5, cbar_kws={'shrink':0.8})
plt.title('Correlation Matrix of Numeric Features', fontsize=16)
plt.tight_layout()
plt.show()

# 9️⃣ Correlation with Price
price_corr = numeric_df.corr()['Price'].drop('Price').sort_values()
plt.figure(figsize=(8,5))
sns.barplot(x=price_corr.index, y=price_corr.values, palette='coolwarm')
plt.title('Correlation of Numeric Features with Price', fontsize=16)
plt.ylabel('Correlation with Price', fontsize=12)
plt.xlabel('Features', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylim(-1,1)
plt.tight_layout()
plt.show()
```

```
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Drop rows with missing GPU
df = df.dropna(subset=['GPU'])

# Convert numeric columns stored as strings
numeric_cols = ['RAM', 'SSD']
for col in numeric_cols:
    # Use raw string for regex and fill NaNs
    df[col] = df[col].astype(str).str.extract(r'(\d+)')[0].fillna(0).astype(int)

# If RAM_Expandable exists
if 'RAM_Expandable' in df.columns:
    df['RAM_Expandable'] = df['RAM_Expandable'].astype(str).str.extract(r'(\d+)')[0].fillna(0).astype(int)

# Encode categorical columns
cat_cols = ['Brand', 'Processor_Brand', 'GPU_Brand', 'Display_type']
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

# Check types and sample
print("Data types after encoding:\n", df.dtypes)
print("\nSample data:\n", df.head())
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# 1️⃣ Full correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='vlag', fmt=".2f", linewidths=0.5, cbar_kws={'shrink':0.8})
plt.title('Correlation Matrix of Numeric Features', fontsize=16)
plt.tight_layout()
plt.show()

# 2️⃣ Correlation of each feature with Price
price_corr = numeric_df.corr()['Price'].drop('Price')  # remove self-correlation

plt.figure(figsize=(8,5))
sns.barplot(x=price_corr.index, y=price_corr.values, palette='coolwarm')
plt.title('Correlation of Numeric Features with Price', fontsize=16)
plt.ylabel('Correlation with Price', fontsize=12)
plt.xlabel('Features', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylim(-1,1)
plt.tight_layout()
plt.show()
```
